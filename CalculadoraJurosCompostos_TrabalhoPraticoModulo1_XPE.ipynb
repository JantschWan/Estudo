```python
#TRABLHAO PRÁTICO DO MÓDULO 1 - CALCULADORA DE JUROS COMPOSTOS - WAN DHER SAAR

# Importação de biblioteca math para cálculo de log
import math as math

def calcular_juros_compostos():
    # Varíaveis que podem ser calculadas
    variaveis = {'vf': None, 'ci': None, 'i': None, 't': None}

    # Mensagem de boas vindas/Escolha de variável
    print("""Bem-vindo à calculadora de Juros Compostos!

O que você gostaria de calcular? (Valor Final 'vf', Capital Inicial 'ci', Período 't' ou Taxa de Juros 'i')?""")

    variavel_escolhida = input().lower()

    while variavel_escolhida not in variaveis:
      print("\nPor favor, escolha entre uma das opções (vf, ci, t ou i)")
      variavel_escolhida = input().lower()

    # Definição dos valores das demais variáveis
    print("\n*Para o valor de i, usar a formatação decimal. Exemplo: 8% = 0.08")
    for var in variaveis:
      if var != variavel_escolhida:
        valor = float(input(f"Qual o valor de {var}:"))
        variaveis[var] = valor

    # Cálculo da variável escolhida
    if variavel_escolhida == 'vf':
        vf = variaveis['ci'] * (1 + variaveis['i'])**variaveis['t']
        print(f"\nO Valor Final 'vf' é R${vf:.2f}")

    elif variavel_escolhida == 'ci':
        ci = variaveis['vf'] / (1 + variaveis['i'])**variaveis['t']
        print(f"\nO capital inicial 'ci' é igual a R${ci:.2f}")

    elif variavel_escolhida == 'i':
        i = (((variaveis['vf'] / variaveis['ci'])**(1 / variaveis['t'])) - 1)
        print(f"\nA taxa de juros 'i' é igual a {i:.2%}")

    elif variavel_escolhida == 't':
        t = math.log(variaveis['vf'] / variaveis['ci']) / math.log(1 + variaveis['i'])
        print(f"\nO período 't' é igual a é {t}")

# Loop principal para permitir ao usuário escolher continuar ou encerrar
while True:
    calcular_juros_compostos()

    resposta = input("\nDeseja calcular novamente? (S para SIM, N para NÃO): \n").upper()
    if resposta != 'S':
        print("Obrigado por usar a calculadora de Juros Compostos. Até mais!")
        break
```
